```python
from news_sentiment import vietnamese_word_tokenize, get_stock_news_data
from trend import get_trend
from utils import LIST_STOCK

import torch
import warnings
import numpy as np
import pandas as pd
from tqdm import tqdm
from datetime import timedelta

from collections import Counter
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import StackingClassifier, RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.pipeline import make_pipeline
```

```python
date_range = pd.date_range(start='2023-03-30', end='2025-03-30', freq='D')

news_by_stock = {stock: {} for stock in list_stock}
corpus = []
i = 0
for stock in list_stock:
    print(stock, i)
    i += 1
    for current_date in tqdm(date_range, desc="Thu thập news", leave=False):
        date_str = current_date.strftime("%Y-%m-%d")
        df = get_stock_news_data(symbol=stock, start_date=date_str, end_date=date_str)
        if df is None or df.empty:
            continue


        df['content_tokenized'] = df['content'].apply(
            lambda t: " ".join(vietnamese_word_tokenize(t))
        )
        df = df[df['content_tokenized'].str.strip().astype(bool)]
        if df.empty:
            continue

        texts = df['content_tokenized'].tolist()
        news_by_stock[stock][date_str] = texts
        corpus.extend(texts)

if not corpus:
    raise RuntimeError("Không tìm thấy tin tức để fit TF-IDF.")

vectorizer = TfidfVectorizer(max_features=5000)
vectorizer.fit(corpus)

test_start = pd.to_datetime("2025-01-30")
test_end   = pd.to_datetime("2025-03-30")

data = {}
for stock, by_date in tqdm(news_by_stock.items(), desc="TF‑IDF & chia train/test", leave=False):
    train_news, train_yesterday, train_week = [], [], []
    test_news,  test_yesterday,  test_week  = [], [], []

    for date_str, texts in by_date.items():

        current_date = pd.to_datetime(date_str)

        tfidf_matrix = vectorizer.transform(texts)
        vectors = tfidf_matrix.toarray().tolist()

        yest = get_trend(
            symbol=stock,
            start_date=(current_date - timedelta(days=1)).strftime("%Y-%m-%d"),
            end_date=date_str,
            based_on='close',
            visualize=False
        )
        if yest is None:
            continue

        nxt = get_trend(
            symbol=stock,
            start_date=date_str,
            end_date=(current_date + timedelta(days=7)).strftime("%Y-%m-%d"),
            based_on='close',
            visualize=False
        )
        if nxt is None:
            continue

        if test_start <= current_date <= test_end:
            test_news.append(vectors)
            test_yesterday.append(yest[2])
            test_week.append(nxt[2])
        else:
            train_news.append(vectors)
            train_yesterday.append(yest[2])
            train_week.append(nxt[2])

    data[stock] = {
        "train": {
            "today_news": train_news,
            "yesterday_trends": train_yesterday,
            "week_trends": train_week,
        },
        "test": {
            "today_news": test_news,
            "yesterday_trends": test_yesterday,
            "week_trends": test_week,
        }
    }

import torch
import torch.nn as nn

torch.save(data, 'tfidf_train_test.pkl')

```

```python
def classification_report_to_latex(report_dict):
    lines = []
    lines.append(r"\begin{table}[H]")
    lines.append(r"\centering")
    lines.append(r"\begin{tabular}{lcccc}")
    lines.append(r"\hline")
    lines.append(r"\textbf{} & \textbf{Precision} & \textbf{Recall} & \textbf{F1-Score} & \textbf{Support} \\")
    lines.append(r"\hline")
    
    # Rows for classes
    for label in ['down', 'sideways', 'up']:
        row = report_dict[label]
        line = f"\\textbf{{{label}}} & {row['precision']:.5f} & {row['recall']:.5f} & {row['f1-score']:.5f} & {int(row['support'])} \\\\"
        lines.append(line)

    lines.append(r"\hline")
    
    # Accuracy
    accuracy = report_dict['accuracy']
    lines.append(f"\\textbf{{accuracy}} & & & {accuracy:.5f} & {int(report_dict['macro avg']['support'])} \\\\")
    
    # Macro avg
    macro = report_dict['macro avg']
    lines.append(f"\\textbf{{macro avg}} & {macro['precision']:.5f} & {macro['recall']:.5f} & {macro['f1-score']:.5f} & {int(macro['support'])} \\\\")
    
    # Weighted avg
    weighted = report_dict['weighted avg']
    lines.append(f"\\textbf{{weighted avg}} & {weighted['precision']:.5f} & {weighted['recall']:.5f} & {weighted['f1-score']:.5f} & {int(weighted['support'])} \\\\")
    
    lines.append(r"\hline")
    lines.append(r"\end{tabular}")
    lines.append(r"\caption{Classification Report}")
    lines.append(r"\end{table}")
    
    latex_code = "\n".join(lines)
    print(latex_code)
```

```python
#naive bayes
warnings.filterwarnings("ignore", category=UserWarning)
label_map = {'down': 0, 'sideways': 1, 'up': 2}

results = {}
all_y_true = []
all_y_pred = []

for symbol, splits in data.items():
    if symbol not in LIST_STOCK:
        continue

    def build_xy(splits_part):
        X, y = [], []
        for news_vecs, trend_vecs, label in zip(
            splits_part["today_news"],
            splits_part["yesterday_trends"],
            splits_part["week_trends"]
        ):
            day_feat     = np.mean(news_vecs, axis=0)
            trend_feat   = np.array(trend_vecs).ravel()
            trend_mapped = np.array([label_map[t] for t in trend_feat])
            feat_vec     = np.concatenate([day_feat, trend_mapped])
            X.append(feat_vec)
            y.append(label)
        return np.vstack(X), np.array(y)

    try:
        X_train, y_train = build_xy(splits["train"])
        X_test,  y_test  = build_xy(splits["test"])
    except ValueError:
        

        print(f"Skipping {symbol}, no data.")
        continue

    clf = MultinomialNB()
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    acc    = accuracy_score(y_test, y_pred)

    report_dict = classification_report(y_test, y_pred, output_dict=True)
    results[symbol] = {
        'accuracy': acc,
        'report'  : report_dict,
        'cm'      : confusion_matrix(y_test, y_pred)
    }

    all_y_true.extend(y_test.tolist())
    all_y_pred.extend(y_pred.tolist())

overall_acc = accuracy_score(all_y_true, all_y_pred)
print(f"\n=== Overall Accuracy: {overall_acc:.4f} ===")
print("=== Overall Classification Report ===")
print(classification_report(all_y_true, all_y_pred))

overall_report = classification_report(all_y_true, all_y_pred, output_dict=True)

f1_list = []
for symbol, res in results.items():
    macro_f1 = res['report']['macro avg']['f1-score']
    f1_list.append((symbol, macro_f1))

f1_list_sorted = sorted(f1_list, key=lambda x: x[1])

lowest_10 = f1_list_sorted[:10]
highest_10 = f1_list_sorted[-10:][::-1]

h = []
l = []
print("\n=== Top 10 symbols by MACRO‑F1 (highest) ===")
for sym, f1 in highest_10:
    h.append(f"{sym} ({f1:.4f})")
print(h)

print("\n=== Top 10 symbols by MACRO‑F1 (lowest) ===")
for sym, f1 in lowest_10:
    l.append(f"{sym} ({f1:.4f})")
print(l)

print(all_y_true, all_y_pred)
classification_report_to_latex(classification_report(all_y_true, all_y_pred, output_dict=True))
```

```python
# Naive Bayes + XGBoost + LightGBM + KNN

import numpy as np
from collections import Counter
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from tensorflow.keras import Sequentiqal # type: ignore
from tensorflow.keras.layers import Dense # type: ignore
from tensorflow.keras.utils import to_categorical # type: ignore


le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc  = le.transform(y_test)
n_classes   = len(le.classes_)   # =3

# 2. Khởi tạo 4 base‑learners
models = [
    GaussianNB(),
    XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'),
    LGBMClassifier(),
    KNeighborsClassifier(n_neighbors=5)
]
n_models = len(models)

# 3. Chuẩn bị meta‑feature containers
meta_train = np.zeros((X_train.shape[0], n_models * n_classes))
meta_test  = np.zeros((X_test.shape[0],  n_models * n_classes))

# 4. K-fold stacking
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for m_idx, model in enumerate(models):
    # lưu preds cho mỗi fold trên test để tính trung bình
    test_fold_preds = []
    for tr_idx, val_idx in skf.split(X_train, y_train_enc):
        X_tr, X_val = X_train[tr_idx], X_train[val_idx]
        y_tr, y_val = y_train_enc[tr_idx], y_train_enc[val_idx]

        model.fit(X_tr, y_tr)
        # predict_proba trả về (n_val, 3)
        proba_val = model.predict_proba(X_val)
        meta_train[val_idx,
                   m_idx*n_classes:(m_idx+1)*n_classes] = proba_val

        # cũng lưu xác suất này cho test
        test_fold_preds.append(model.predict_proba(X_test))

    # lấy trung bình preds trên 5 fold cho tập test
    meta_test[:, m_idx*n_classes:(m_idx+1)*n_classes] = np.mean(test_fold_preds, axis=0)

# 5. Xây dựng và train MLP meta‑learner
y_train_ohe = to_categorical(y_train_enc, num_classes=n_classes)
y_test_ohe  = to_categorical(y_test_enc,  num_classes=n_classes)

nn = Sequential([
    Dense(32, input_shape=(n_models*n_classes,), activation='relu'),
    Dense(16, activation='relu'),
    Dense(n_classes, activation='softmax')
])
nn.compile(optimizer='adam',
           loss='categorical_crossentropy',
           metrics=['accuracy'])

nn.fit(meta_train, y_train_ohe,
       epochs=500, batch_size=32,
       validation_split=0.1,
       verbose=2)

# 6. Đánh giá
preds = np.argmax(nn.predict(meta_test), axis=1)
print("Stacking accuracy:", accuracy_score(y_test_enc, preds))
```
