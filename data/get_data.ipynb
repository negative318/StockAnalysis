```python
import requests
import html
import json

results = {}
list_stock = ["VCB"]
for symbol in list_stock:
    limit = 10
    offset = 0
    sort = 1

    url = f'https://api.fireant.vn/symbols/{symbol}/posts?type=1&limit={limit}&offset={offset}&sort={sort}'
    headers = {
        "Authorization": f"Bearer {access_token}"
    }
    response = requests.get(url, headers=headers, timeout=10)
    if response.status_code == 200:
        data = response.json()
        print(f"Start {symbol}")
        news_by_day = {}
        for post in data:
            date_str = post.get("date")
            day = date_str.split('T')[0] if date_str else "Unknown"

            url_content = f"https://api.fireant.vn/posts/{post.get("postID")}"
            response_content = requests.get(url_content, headers=headers, timeout=10).json().get("content")
            content = html.unescape(response_content)
            post_data = {
                'date': post.get("date"),
                'title': post.get("title"),
                'description': post.get("description"),
                'content': content,
                'taggedSymbols': post.get("taggedSymbols"),

            }
            if day not in news_by_day:
                news_by_day[day] = []
            news_by_day[day].append(post_data)
        
        path = f"data/{symbol}/news.json"
        data_to_save = {
            "symbol": symbol,
            "news": news_by_day
        }
    print(json.dumps(news_by_day, indent=2, ensure_ascii=False))
    #     with open(path, 'w', encoding='utf-8') as f:
    #         json.dump(data_to_save, f, ensure_ascii=False, indent=4)
    #     print(f"Write {symbol}, to: {path}")
    # else:
    #     print(f"Yu cu tht bi cho {symbol}, m li: {response.status_code}")
    #     results[symbol] = {"error": response.text}
```

```python
import pandas as pd
all_articles = []
for day, articles in news_by_day.items():
    for art in articles:
        # nu mun gi thm ct 'day' (ngy gc), c th:
        art_with_day = {"day": day, **art}
        all_articles.append(art_with_day)

# 2. Dng json_normalize  flat JSON thnh bng
df = pd.json_normalize(all_articles)

# 3. Chn cc ct bn cn hin th
df['taggedSymbol'] = df['taggedSymbols'].apply(
    lambda lst: lst[0]['symbol'] if isinstance(lst, list) and len(lst) > 0 else None
)

from IPython.display import display, HTML

# Hin th DataFrame di dng HTML  xem ton b ni dung
display(HTML(df[['content']].to_html()))
```
