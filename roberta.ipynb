```python
from news_sentiment import vietnamese_word_tokenize, get_stock_news_data
from trend import get_trend
from utils import LIST_STOCK

from transformers import AutoTokenizer, AutoModel

import torch
import warnings
import numpy as np
import pandas as pd
from tqdm import tqdm
from datetime import timedelta
from collections import Counter


from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.pipeline import make_pipeline
```

```python
date_range = pd.date_range(start='2023-03-30', end='2025-03-30', freq='D')
news_by_stock = {stock: {} for stock in list_stock}
corpus = []

for stock in list_stock:
    for current_date in tqdm(date_range, desc=f"Thu thập news {stock}", leave=False):
        date_str = current_date.strftime("%Y-%m-%d")
        df = get_stock_news_data(symbol=stock, start_date=date_str, end_date=date_str)
        if df is None or df.empty:
            continue

        # Tokenize tiếng Việt
        df['content_tokenized'] = df['content'].apply(
            lambda t: " ".join(vietnamese_word_tokenize(t))
        )
        df = df[df['content_tokenized'].str.strip().astype(bool)]
        if df.empty:
            continue

        texts = df['content_tokenized'].tolist()
        news_by_stock[stock][date_str] = texts
        corpus.extend(texts)

if not corpus:
    raise RuntimeError("Không tìm thấy tin tức để xử lý embedding.")


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# PhoGPT-base base model
tokenizer = AutoTokenizer.from_pretrained("vinai/phobert-base")
model     = AutoModel.from_pretrained("vinai/phobert-base")
model.to(device)
model.eval()


test_start = pd.to_datetime("2025-01-30")
test_end   = pd.to_datetime("2025-03-30")

data = {}
i = 0
for stock, by_date in tqdm(news_by_stock.items(), desc="TF‑GPT & chia train/test", leave=False):
    print(stock, i)
    i+= 1
    train_news, train_yest, train_week = [], [], []
    test_news,  test_yest,  test_week  = [], [], []

    for date_str, texts in by_date.items():
        current_date = pd.to_datetime(date_str)
        if not texts:
            continue

        enc = tokenizer(texts, return_tensors="pt", padding=True, truncation=True, max_length=128)
        enc = {k: v.to(device) for k, v in enc.items()}
        with torch.no_grad():
            out = model(**enc)
            embeddings = out.last_hidden_state.mean(dim=1)
        embeddings = embeddings.cpu().numpy()
        # embeddings = embeddings.mean(axis = 0)


        yest = get_trend(
            symbol=stock,
            start_date=(current_date - timedelta(days=1)).strftime("%Y-%m-%d"),
            end_date=date_str,
            based_on='close',
            visualize=False
        )
        if yest is None:
            continue


        nxt = get_trend(
            symbol=stock,
            start_date=date_str,
            end_date=(current_date + timedelta(days=7)).strftime("%Y-%m-%d"),
            based_on='close',
            visualize=False
        )
        if nxt is None:
            continue

        # Phân bổ vào train hoặc test
        if test_start <= current_date <= test_end:
            test_news.append(embeddings.tolist())
            test_yest.append(yest[2])
            test_week.append(nxt[2])
        else:
            train_news.append(embeddings.tolist())
            train_yest.append(yest[2])
            train_week.append(nxt[2])

    data[stock] = {
        "train": {
            "today_news": train_news,
            "yesterday_trends": train_yest,
            "week_trends": train_week,
        },
        "test": {
            "today_news": test_news,
            "yesterday_trends": test_yest,
            "week_trends": test_week,
        }
    }


torch.save(data, 'phogpt_train_test.pkl')
print("Done. Saved to phogpt_train_test.pkl")
```

```python
def classification_report_to_latex(report_dict):
    lines = []
    lines.append(r"\begin{table}[H]")
    lines.append(r"\centering")
    lines.append(r"\begin{tabular}{lcccc}")
    lines.append(r"\hline")
    lines.append(r"\textbf{} & \textbf{Precision} & \textbf{Recall} & \textbf{F1-Score} & \textbf{Support} \\")
    lines.append(r"\hline")
    
    # Rows for classes
    for label in ['down', 'sideways', 'up']:
        row = report_dict[label]
        line = f"\\textbf{{{label}}} & {row['precision']:.5f} & {row['recall']:.5f} & {row['f1-score']:.5f} & {int(row['support'])} \\\\"
        lines.append(line)

    lines.append(r"\hline")
    
    # Accuracy
    accuracy = report_dict['accuracy']
    lines.append(f"\\textbf{{accuracy}} & & & {accuracy:.5f} & {int(report_dict['macro avg']['support'])} \\\\")
    
    # Macro avg
    macro = report_dict['macro avg']
    lines.append(f"\\textbf{{macro avg}} & {macro['precision']:.5f} & {macro['recall']:.5f} & {macro['f1-score']:.5f} & {int(macro['support'])} \\\\")
    
    # Weighted avg
    weighted = report_dict['weighted avg']
    lines.append(f"\\textbf{{weighted avg}} & {weighted['precision']:.5f} & {weighted['recall']:.5f} & {weighted['f1-score']:.5f} & {int(weighted['support'])} \\\\")
    
    lines.append(r"\hline")
    lines.append(r"\end{tabular}")
    lines.append(r"\caption{Classification Report}")
    lines.append(r"\end{table}")
    
    latex_code = "\n".join(lines)
    print(latex_code)
```

```python
#naive bayes
warnings.filterwarnings("ignore", category=UserWarning)
label_map = {'down': 0, 'sideways': 1, 'up': 2}

results = {}
all_y_true = []
all_y_pred = []

for symbol, splits in data.items():
    if symbol not in LIST_STOCK:
        continue

    def build_xy(splits_part):
        X, y = [], []
        for news_vecs, trend_vecs, label in zip(
            splits_part["today_news"],
            splits_part["yesterday_trends"],
            splits_part["week_trends"]
        ):
            day_feat     = np.mean(news_vecs, axis=0)
            trend_feat   = np.array(trend_vecs).ravel()
            trend_mapped = np.array([label_map[t] for t in trend_feat])
            feat_vec     = np.concatenate([day_feat, trend_mapped])
            X.append(feat_vec)
            y.append(label)
        return np.vstack(X), np.array(y)

    try:
        X_train, y_train = build_xy(splits["train"])
        X_test,  y_test  = build_xy(splits["test"])
    except ValueError:
        

        print(f"Skipping {symbol}, no data.")
        continue

    clf = GaussianNB()
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    acc    = accuracy_score(y_test, y_pred)

    report_dict = classification_report(y_test, y_pred, output_dict=True)
    results[symbol] = {
        'accuracy': acc,
        'report'  : report_dict,
        'cm'      : confusion_matrix(y_test, y_pred)
    }

    all_y_true.extend(y_test.tolist())
    all_y_pred.extend(y_pred.tolist())

overall_acc = accuracy_score(all_y_true, all_y_pred)
print(f"\n=== Overall Accuracy: {overall_acc:.4f} ===")
print("=== Overall Classification Report ===")
print(classification_report(all_y_true, all_y_pred))

overall_report = classification_report(all_y_true, all_y_pred, output_dict=True)

f1_list = []
for symbol, res in results.items():
    macro_f1 = res['report']['macro avg']['f1-score']
    f1_list.append((symbol, macro_f1))

f1_list_sorted = sorted(f1_list, key=lambda x: x[1])

lowest_10 = f1_list_sorted[:10]
highest_10 = f1_list_sorted[-10:][::-1]

h = []
l = []
print("\n=== Top 10 symbols by MACRO‑F1 (highest) ===")
for sym, f1 in highest_10:
    h.append(f"{sym} ({f1:.4f})")
print(h)

print("\n=== Top 10 symbols by MACRO‑F1 (lowest) ===")
for sym, f1 in lowest_10:
    l.append(f"{sym} ({f1:.4f})")
print(l)

print(all_y_true, all_y_pred)
classification_report_to_latex(classification_report(all_y_true, all_y_pred, output_dict=True))
```

```python
# AdaBoost
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

# Bản đồ nhãn
label_map = {'down': 0, 'sideways': 1, 'up': 2}

results = {}
all_y_true = []
all_y_pred = []

for symbol, splits in data.items():
    if symbol not in LIST_STOCK:
        continue

    def build_xy(splits_part):
        X, y = [], []
        for news_vecs, trend_vecs, label in zip(
                splits_part["today_news"],
                splits_part["yesterday_trends"],
                splits_part["week_trends"]
        ):
            day_feat     = np.mean(news_vecs, axis=0)
            trend_feat   = np.array(trend_vecs).ravel()
            trend_mapped = np.array([label_map[t] for t in trend_feat])
            feat_vec     = np.concatenate([day_feat, trend_mapped])
            X.append(feat_vec)
            y.append(label_map[label])
        return np.vstack(X), np.array(y)

    try:
        X_train, y_train = build_xy(splits["train"])
        X_test,  y_test  = build_xy(splits["test"])
    except ValueError:
        print(f"Skipping {symbol}, no data.")
        continue

    unique_classes = np.unique(y_train)
    if len(unique_classes) < 2:
        print(f"Skipping {symbol}, chỉ có {len(unique_classes)} lớp trong training.")
        continue

    missing = set(label_map.values()) - set(unique_classes)
    if missing:
        print(f"Thêm mẫu giả cho các lớp thiếu: {missing} ở symbol {symbol}")
        n_features = X_train.shape[1]
        for cls in missing:
            X_train = np.vstack([X_train, np.zeros((1, n_features))])
            y_train = np.append(y_train, cls)

    labels = [0, 1, 2]
    target_names = ['down', 'sideways', 'up']

    clf = make_pipeline(
        StandardScaler(),
        AdaBoostClassifier(
            # số vòng boosting; có thể tăng lên để cải thiện độ chính xác
            n_estimators=200,
            # tốc độ học; giá trị mặc định là 1.0
            learning_rate=1.0,
            # sử dụng DecisionTreeClassifier làm base estimator
            # với độ sâu tối đa 1 (stump) – theo mặc định
            random_state=42
        )
    )
    clf.fit(X_train, y_train)

    y_pred = clf.predict(X_test)
    acc    = accuracy_score(y_test, y_pred)

    report_dict = classification_report(y_test, y_pred, output_dict=True)
    results[symbol] = {
        'accuracy': acc,
        'report': report_dict,
        'cm': confusion_matrix(y_test, y_pred)
    }

    all_y_true.extend(y_test.tolist())
    all_y_pred.extend(y_pred.tolist())

overall_acc = accuracy_score(all_y_true, all_y_pred)
print(f"\n=== Overall Accuracy: {overall_acc:.4f} ===")
print("=== Overall Classification Report ===")
print(classification_report(all_y_true, all_y_pred))

overall_report = classification_report(
    all_y_true,
    all_y_pred,
    labels=labels,
    target_names=target_names,
    output_dict=True
)

f1_list = [
    (sym, res['report']['macro avg']['f1-score'])
    for sym, res in results.items()
]
f1_sorted = sorted(f1_list, key=lambda x: x[1])

lowest_10 = f1_sorted[:10]
highest_10 = f1_sorted[-10:][::-1]

h = []
l = []
print("\n=== Top 10 symbols by MACRO‑F1 (highest) ===")
for sym, f1 in highest_10:
    h.append(f"{sym} ({f1:.4f})")
print(h)

print("\n=== Top 10 symbols by MACRO‑F1 (lowest) ===")
for sym, f1 in lowest_10:
    l.append(f"{sym} ({f1:.4f})")
print(l)

label_map = {0:'down', 1:'sideways', 2:'up'}
all_y_true_mapped =  [label_map[t] for t in all_y_true]
all_y_pred_mapped =  [label_map[t] for t in all_y_pred]
print(all_y_pred_mapped, all_y_true_mapped)

classification_report_to_latex(classification_report(
    all_y_true,
    all_y_pred,
    labels=labels,
    target_names=target_names,
    output_dict=True
))
```
